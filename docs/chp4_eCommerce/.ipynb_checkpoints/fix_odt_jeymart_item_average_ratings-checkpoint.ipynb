```python
import elasticsearch;
import math;

def _getAllItems(_es):
    _res = _es.search(
        index='jeymart_item',
        doc_type='doc',
        body=
        {
            "size": 5000, 
            "query": {
                "match_all": {}
            }
        }
    );
    return _res['hits']['hits'];

def _prepareBulkEntries(_items, _stockCode): 
    _entries=[];
    for _item in _items:
        _id = _item['_id'];
        # add meta
        _entries.push({ "update" : { "_index" : "jeymart_user_trx", "_type" : "doc", "_id" : _id } });
        _entries.push({ "doc" : {"k_stock_code" :  _stockCode } });
    return _entries;
    

def _prepareUpdateEntries(_es, _items):
    
    for _item in _items:
        _id = _item['_id']
        _desc = _item['_source']['t_description']
        
        # query for the average...
        _res=_es.search(
            index='jeymart_user_trx',
            doc_type='doc',
            body=
            {
              "size": 10000, 
              "query": {
                "match_phrase": {
                  "t_description": _desc
                }
              }
            }            
        );
        # build the bulk update entries
        
        print(len(_res['hits']['hits']));
        
_es = elasticsearch.Elasticsearch([ "http://localhost:9201"], requestTimeout="600000" );
# get back all items
_items=_getAllItems(_es);
_prepareUpdateEntries(_es, _items);



# add back the "k_stock_code" and the overall average of "i_ratings" for jeymart_item


```
